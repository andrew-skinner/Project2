```python
#Dependencies
import pandas as pd
import numpy as np
import requests
import json
import pprint
```

```python
#The base url indicates the dataset with "WWDI"
base_url = "https://www.quandl.com/api/v3/datasets/WWDI/"
```

```python
api_key = "&api_key=T67YrzQjRU2zXJh_s9tV"
```

```python
country_codes = pd.read_csv("country_codes.csv", delimiter="|", error_bad_lines=False)
country_codes.head()
```

```python
population = {}
for index, row in country_codes.iterrows():
    try:
        var = row["CODE"]
        country = row["COUNTRY"]
        dataset = "_SP_POP_TOTL"
        query = base_url + var + dataset + "?&" + api_key
        response = requests.get(query)
        data = response.json()
        population.update({country: (data['dataset']['data'])[0][1]})
    except KeyError:
        continue
population
```

```python
df = pd.DataFrame.from_dict(population, orient='index')
```
