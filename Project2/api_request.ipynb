```python
#Pulling in the CSV file with countries
country_codes = pd.read_csv("country_codes.csv", delimiter="|", error_bad_lines=False)
```

```python
Population= {}

for index, row in country_codes.iterrows():
    try:
        var = row["CODE"]
        country = row["COUNTRY"]
        dataset = "_SP_POP_TOTL"
        query = base_url + var + dataset + "?&" + api_key
        response = requests.get(query)
        data = response.json()
        Population.update({country: (data['dataset']['data'])[0][1]})
    except KeyError:
        continue

#Conversion into the DataFrame
#NOTE that we are using the "FROM_DICT" to properly bring in the data
DF0 = pd.DataFrame.from_dict(Population, orient='index')
```
